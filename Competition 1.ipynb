```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E:\hotel_bookings.csv")
```

```python
df.describe()
```

```python
df.head()
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
df.info()
```

```python
X = df.drop(['is_canceled','company'], axis = 1)
y = df['is_canceled']
X["country"].fillna(X.country.mode(), inplace = True)
X["agent"].fillna(9, inplace = True)
X["children"].fillna(0, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X.head()
```

```python
X_numeric = list(X.dtypes[X.dtypes != "object"].index)
X[X_numeric].head()
```

```python
X[X_numeric].describe()
```

```python
sns.heatmap(X[X_numeric].isnull(), cbar=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model = RandomForestClassifier(n_estimators=100, oob_score=True, random_state=42)
```

```python
model.fit(X_train[X_numeric],y_train)
```

```python
y_pred = model.predict(X_test[X_numeric])
feature_importance = pd.Series(model.feature_importances_, index = X[X_numeric].columns)
feature_importance.sort_values().plot(kind="barh")
model.feature_importances_
```

```python
accuracy_score(y_test,y_pred)
```

```python
X_category = X.drop(X[X_numeric], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
y_pred = model.predict(X_test)
feature_importance = pd.Series(model.feature_importances_, index = X.columns)
feature_importance.sort_values().plot(kind="barh")
```

```python
accuracy_score(y_test,y_pred)
```
